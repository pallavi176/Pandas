### 1 . how many campaign available in this dataset .

### 2 . how many users do we have with housing and personal loan . 

```python
bank[(bank['housing'] == 'yes') & (bank["loan"] == 'yes')]
```

### 3. how many person do we have whose age is 60+ . 

```python
len(bank[bank['age'] >= 60])
```

### 4 . in which month we have trageted most of the customer .

```python
bank.groupby(['month','y']).agg({'count'})
```

```python
bank.groupby(['month','y']).size().reset_index().groupby('y')[[0]].max().iloc[1]
```

### 5 . which mode of call is giving you more result

```python
bank.groupby(['contact','y']).count()
```

```python
bank.groupby(['contact','y']).size().reset_index().groupby('y')[[0]].max().loc['yes']
```

```python
bank.groupby(['contact','y']).size().reset_index()
```
